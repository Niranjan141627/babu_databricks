```python
%sql
-- Dimension: Customer
CREATE TABLE DimCustomer (
    customer_key INT PRIMARY KEY,
    customer_id VARCHAR(10),    -- Business/Natural key
    customer_name VARCHAR(50),
    city VARCHAR(50)
);

-- Dimension: Product
CREATE TABLE DimProduct (
    product_key INT PRIMARY KEY,
    product_id VARCHAR(10),     -- Business/Natural key
    product_name VARCHAR(50),
    category VARCHAR(50)
);
```

```python
%sql
-- Fact: Sales
CREATE TABLE FactSales (
    sales_key INT PRIMARY KEY,
    customer_key INT,       -- Foreign Key to DimCustomer
    product_key INT,        -- Foreign Key to DimProduct
    sales_date DATE,
    quantity INT,
    total_amount DECIMAL(10,2),
    FOREIGN KEY (customer_key) REFERENCES DimCustomer(customer_key),
    FOREIGN KEY (product_key) REFERENCES DimProduct(product_key)
);
```
